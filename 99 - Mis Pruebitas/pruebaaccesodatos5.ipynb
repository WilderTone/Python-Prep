```python
import pandas as pd
import numpy as np

#Creando DATAFRAME desde un ARCHIVO csV

df_boston= pd.read_csv("Boston House Prices.csv")

df_boston
```

```python
##REgresion Lineal Multiple

import statsmodels.api as sm

y=df_boston['Value']  # variable dependiente Axis Y
X=df_boston[['Rooms','Distance']]  # variables  independiente o predictor axis X


X=sm.add_constant(X) ##Agregando la Constante

## Ajustamdo el modelo
lm = sm.OLS(y,X).fit() ## lm lineal model  metodo para ajustar

lm.summary()
```
