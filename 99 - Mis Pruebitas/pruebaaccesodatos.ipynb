```python
#  A Agregando columna desde un array , creando el array

import numpy as np

language_score= np.arange(0,1000) # creando un array con valores de 0 a 1000
len(language_score)
```

```python
# B Agregando columna desde un array , pasando los valores del array al dataframe

df_exams['languaje score']=language_score
df_exams
```

```python
# Usando numpy para crear numeros aleatorios

languaje_score2 = np.random.randint(1,100, size=1000) # creando array con valores aleatorios enteros
# Observacion
# en la funcion np.random.randint(1,100, size=1000) el valor minimo es inclusivo y el valor maximo es exclusivo
# quiere decir que en los resultados obtenidos en el array el valor minimo puede que aparezca pero el valor maximo NO va a estar

df_exams['languaje score']=languaje_score2 #pasando los valores al dataframe
df_exams
```

```python

#Valor minimo inclusivo y Valor maximo exclusivo
min(languaje_score2)
max(languaje_score2)
```
